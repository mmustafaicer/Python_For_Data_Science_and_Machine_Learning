## Method Used

K Means Clustering is an [**unsupervised learning**](https://en.wikipedia.org/wiki/Unsupervised_learning) algorithm that tries to cluster data based on their similarity. Unsupervised learning means that there is no outcome to be predicted, and the algorithm just tries to find patterns in the data. In k means clustering, we have the specify the number of clusters we want the data to be grouped into. The algorithm randomly assigns each observation to a cluster, and finds the centroid of each cluster. Then, the algorithm iterates through two steps:
Reassign data points to the cluster whose centroid is closest. Calculate new centroid of each cluster. These two steps are repeated till the within cluster variation cannot be reduced any further. The within cluster variation is calculated as the sum of the euclidean distance between the data points and their respective cluster centroids.

## Import Libraries

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

## Create some Data

`make_blobs()` method is used to create artificial cluster so we can do a cluster analysis on it.

```python
from sklearn.datasets import make_blobs
```

`n_samples` = number of observations<br>
`n_features` = number of columns/variables<br>
`centers` = how many centers or clusters there will be

```python
# Create Data
data = make_blobs(n_samples=200, n_features=2, 
                           centers=4, cluster_std=1.8,random_state=101)
```

If you look our `data`, it is a tuple.

```python
data[1] # for features
```

```python
data[0].shape # 200 observations and 2 columns
```

```python
data[1] # centers
```

## Visualize Data

Plot first column agains the second column.

`c=data[1]` refers to centers or number of clusters.

```python
plt.scatter(data[0][:,0],data[0][:,1],c=data[1],cmap='rainbow')
```

## Creating the Clusters

```python
from sklearn.cluster import KMeans
```

Instantiate our KMeans model.

```python
kmeans = KMeans(n_clusters=4)
```

```python
kmeans.fit(data[0])
```

```python
kmeans.cluster_centers_
```

```python
kmeans.labels_
```

Let's plot our K-means model with original dataset plot. So we can see how it performed.

```python
f, (ax1, ax2) = plt.subplots(1, 2, sharey=True,figsize=(10,6))
ax1.set_title('K Means')
ax1.scatter(data[0][:,0],data[0][:,1],c=kmeans.labels_,cmap='rainbow')
ax2.set_title("Original")
ax2.scatter(data[0][:,0],data[0][:,1],c=data[1],cmap='rainbow')
```

You should note, the colors are meaningless in reference between the two plots.

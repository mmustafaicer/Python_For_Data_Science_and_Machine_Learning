Import` matplotlib.pyplot` as plt and make `%matplotlib inline`.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

Create our data.

```python
x = np.arange(0,100)
y = x*2
z = x**2
```

## Question 1
Create a figure and plot the data.

```python
# create a figure object
fig = plt.figure()

#plot the data
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, "r--", lw=2)
ax.set_xlabel("X Values")
ax.set_ylabel("Y Values")
ax.set_title("My Plot")
```

## Question 2
Create an empty canvas and put two axes within the same figure.

```python
fig = plt.figure()

# Larger plot
ax1 = fig.add_axes([0,0,1,1])
ax1.plot(x,y, "r", lw=2)
ax1.set_xlabel("X Values")
ax1.set_ylabel("Y Values")
ax1.set_title("Larger Plot")

# Smaller plot
ax2 = fig.add_axes([0.2,0.6,0.2,0.2])
ax2.plot(x,z, "b", lw=2)
ax2.set_xlabel("X Values")
ax2.set_ylabel("Z Values")
ax2.set_title("Smaller Plot")
```

## Question 3
Use `plt.subplots()` function to create two plots side by side in one row.

```python
fig,axes = plt.subplots(1,2,figsize=(16,3))

axes[0].plot(x,y, "r",lw=3, marker="o")
axes[0].set_xlabel("X Values")
axes[0].set_ylabel("Y Values")
axes[0].set_title("X versus Y")

axes[1].plot(x,z, "b--", lw=2, marker="+")
axes[1].set_xlabel("X Values")
axes[1].set_ylabel("Z Values")
axes[1].set_title("X versus Z")
```
